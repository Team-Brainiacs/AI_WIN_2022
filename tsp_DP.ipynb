**Dynamic Programming** :

It is a deterministic optimization technique that is used to solve problems by breaking them down into smaller subproblems and solving them recursively. The key idea behind dynamic programming is to store the solutions to subproblems in a table (known as a memoization table) so that they can be reused later. This way, the same subproblem does not need to be solved multiple times, which reduces the time complexity of the algorithm(less than n! but still exponential).

```python
import sys

# Function to find the minimum cost to travel all cities and the path
def TSP(points, current_city, visited, cost, path):
    # If all cities are visited, add the starting city to the path and return cost
    if visited == (1<<len(points))-1:
        path.append(current_city)
        return cost + dist(current_city, 0), path

    # Initialize minimum cost as infinity and an empty path
    min_cost = sys.maxsize
    min_path = []

    # Try all unvisited cities as next move
    for city in range(len(points)):
        if visited & (1<<city) == 0:
            # Add the current city to the path
            new_path = path + [current_city]
            # Find the cost and path of the next move
            new_cost, path_out = TSP(points, city, visited|(1<<city), cost + dist(current_city, city), new_path)
            # Update the minimum cost and path if a better solution is found
            if new_cost < min_cost:
                min_cost = new_cost
                min_path = path_out

    return min_cost, min_path

# Function to calculate distance between two cities
def dist(city1, city2):
    x1, y1 = points[city1]
    x2, y2 = points[city2]
    return ((x1-x2)**2 + (y1-y2)**2)**0.5

# Driver code
if __name__ == "__main__":
    points = [(0,0), (1,1), (2,2), (3,3)]
    start_city = 0
    visited = 1
    cost = 0
    path = [0]
    min_cost, min_path = TSP(points, start_city, visited, cost, path)
    print("Minimum cost to travel all cities:", min_cost)
    print("Path:", min_path)
```

**FAQs**

1) How is DP a deterministic algorithm?

Ans. Dynamic programming is deterministic because it always produces the same result for the same input. It does not rely on any randomness or chance, and the solutions to the subproblems are computed using a set of fixed rules and formulas.
